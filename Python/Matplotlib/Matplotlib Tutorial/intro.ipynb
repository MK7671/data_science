### Scatter Plots

```python
x = [1,2,3,4,5,6,7,8]
y = [5,2,4,2,1,4,5,2]
plt.scatter(x,y, label='skitscat', color='k', marker='x', s=500)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
```

### Stack Plots

```python
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

# linewidth: change the size of line in legend
plt.plot([], [], color='m', label='sleeping', linewidth=5)
plt.plot([], [], color='c', label='eating', linewidth=5)
plt.plot([], [], color='r', label='working', linewidth=5)
plt.plot([], [], color='k', label='playing', linewidth=5)

# x and y's elements
plt.stackplot(days, sleeping, eating, working, playing, colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Interesting Graph\nCheck it out')
plt.legend()
```

### Pie Charts

```python
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
cols = ['c','m','r','y']

# startangle: give angle at the 12'o clock way
# shadow: add shadow
# explode:
# autopct: show percentage
plt.pie(slices, 
        labels=activities, 
        colors = cols, 
        startangle=90, 
        shadow=True, 
        explode=(0,0.1,0,0),
        autopct="%1.1f%%")

plt.title('Interesting Graph\nCheck it out')
```
